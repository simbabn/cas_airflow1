```python
import requests
import pandas as pd
import json

url_api = "https://api.bouyguestelecom.fr/ventes/produits?type=phone"
header = {'Content-Type': 'application/json', 'x-version': '4'}

req = requests.get(url_api, headers=header)
req_json = req.json()
data = req_json["produits"]

df = pd.DataFrame(data).T

colonnes = df.columns
print(colonnes)
```
